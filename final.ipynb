```python
train_cat=data.select_dtypes(include='object')
train_cat
```

```python
def mis_cat(series):
    mode=series.value_counts().index[0]
    series=series.fillna(mode)
    return series
```

```python
for x in train_cat:
    train_cat[x]=mis_cat(train_cat[x])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import sklearn.metrics as metrics
from sklearn.metrics import confusion_matrix
import sklearn.preprocessing as pre
import sklearn.model_selection as ms
```

```python
train_cat.isnull().sum()
```

```python
X0=data
```

```python
X0
```

```python
X0.dtypes.value_counts()
```

```python
col_ob = X0.select_dtypes(include = 'object').columns
for x in col_ob:
    X0[x] = pre.LabelEncoder().fit_transform(X0[x])
```

```python
X0.dtypes.value_counts()
```

```python
X0.dtypes
```

```python
X = X0.drop('response_binary',axis=1)
Y = X0['response_binary']
```

```python
X.job.unique()
```

```python
data.job.unique()
```

```python
X.head()
```

```python
Y.head()
```

```python
X
```

```python
test_size= 0.20
seed = 10
X_train, X_test, Y_train, Y_test= ms.train_test_split(X, Y, test_size=test_size, random_state=seed)
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
LR = LogisticRegression()
LR.fit(X_train, Y_train)
```

```python
predictions = LR.predict(X_test)
```

```python
import pylab as pl

print(confusion_matrix(Y_test, predictions))

cm = confusion_matrix(Y_test, predictions)
pl.matshow(cm)
pl.title('Confusion matrix of the classifier')
pl.colorbar()
pl.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test, predictions))
```

```python
print("Accuracy:", metrics.accuracy_score(Y_test, predictions))
print("Precision:", metrics.precision_score(Y_test, predictions))
```
